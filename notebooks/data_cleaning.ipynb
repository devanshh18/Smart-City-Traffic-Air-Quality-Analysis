```python
# importing required libraries
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
engine = create_engine("mysql+pymysql://root:root@localhost/smart_city_analysis")
```

```python
traffic_df = pd.read_sql("SELECT * FROM traffic_data", engine)
```

```python
air_df = pd.read_sql("SELECT * FROM air_quality_data", engine)
```

```python
traffic_df.head()
```

```python
traffic_df.info()
```

```python
air_df.info()
```

```python
# checking missing values
traffic_df.isnull().sum()
```

```python
air_df.isnull().sum()
```

```python
# drop duplicates
traffic_df = traffic_df.drop_duplicates()
air_df = air_df.drop_duplicates()
```

```python
traffic_df.duplicated().sum()
```

```python
air_df.duplicated().sum()
```

```python
# converting traffic timestamp to datetime
traffic_df['timestamp'] = pd.to_datetime(traffic_df['timestamp'])
```

```python
traffic_df.dtypes
```

```python
# converting air quality date column
air_df['last_update'] = pd.to_datetime(
    air_df['last_update'],
    dayfirst=True,
    errors='coerce'
)
```

```python
air_df.dtypes
```

```python
# converting pollutant columns to numeric
air_df['pollutant_min'] = pd.to_numeric(air_df['pollutant_min'], errors='coerce')
air_df['pollutant_max'] = pd.to_numeric(air_df['pollutant_max'], errors='coerce')
air_df['pollutant_avg'] = pd.to_numeric(air_df['pollutant_avg'], errors='coerce')
```

```python
air_df.dtypes
```

```python
# handling missing pollutant values
air_df[['pollutant_min', 'pollutant_max', 'pollutant_avg']].isnull().sum()
```

```python
air_df.isnull().sum()
```

```python
# replacing the missing values with median
air_df['pollutant_min'] = air_df['pollutant_min'].fillna(air_df['pollutant_min'].median())
air_df['pollutant_max'] = air_df['pollutant_max'].fillna(air_df['pollutant_max'].median())
air_df['pollutant_avg'] = air_df['pollutant_avg'].fillna(air_df['pollutant_avg'].median())
```

```python
air_df[['pollutant_min','pollutant_max','pollutant_avg']].isnull().sum()
```
